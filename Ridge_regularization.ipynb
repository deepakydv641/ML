```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_diabetes
```

```python
x,y=load_diabetes(return_X_y=True)
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
res=lr.predict(x_test)
```

```python
from sklearn.metrics import r2_score,mean_squared_error
r2_score(y_test,res)
```

```python
np.sqrt(mean_squared_error(y_test,res))
```

Now i will do the same as this i have done before by using Ridge regularization to see the difference

```python
from sklearn.linear_model import Ridge
```

```python
ri=Ridge(alpha=100000)
ri.fit(x_train,y_train)
```

```python
result=ri.predict(x_test)
```

```python
r2_score(y_test,result)
```

```python
np.sqrt(mean_squared_error(y_test,result))
```

if r2_score is negative it means that it goes to underfitting ,if r2_score is -ve means y=mean would have performed bettre than your curreint model
