```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.impute import MissingIndicator,SimpleImputer
```

```python
path=r"C:\Users\deepa\Downloads\train.csv"
df=pd.read_csv(path,usecols=['Age','Fare','Survived'])
df.head()
```

```python
df.isna().mean()*100
```

```python
x=df.drop(columns=['Survived'])
y=df['Survived']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
y_test.head()
```

```python
si=SimpleImputer()
x_train_si=si.fit_transform(x_train)
x_test_si=si.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
lr.fit(x_train_si,y_train)
```

```python
result=lr.predict(x_test_si)
```

```python
result
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,result)
```

```python
## Now i will do same thing but using Missing Indicator
mi=MissingIndicator()
x_train_mi=mi.fit_transform(x_train)
x_test_mi=mi.transform(x_test)
```

```python
x_test_mi
```

```python
x_train['age_na']=x_train_mi
```

```python
x_test['age_na']=x_test_mi
```

```python
x_test
```

```python
si=SimpleImputer()
x_train_trf=si.fit_transform(x_train)
x_test_trf=si.transform(x_test)
```

```python
lr=LogisticRegression()
lr.fit(x_train_trf,y_train)
```

```python
res=lr.predict(x_test_trf)
```

```python
accuracy_score(y_test,res)
```
