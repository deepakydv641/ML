```python
import numpy as np
import pandas as pd
```

```python
path=r"C:\Users\deepa\Downloads\train.csv"
d=pd.read_csv(path)
```

```python
d.head()
```

```python
df=d.drop(columns=['PassengerId','Name','Ticket','Cabin'])
```

```python
df.head()
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
from sklearn.impute import SimpleImputer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.pipeline import Pipeline,make_pipeline
from sklearn.feature_selection import SelectKBest,chi2
from sklearn.tree import DecisionTreeClassifier
```

```python
df.isna().sum()
```

```python
# Step->1 = Train Test Split
x_train,x_test,y_train,y_test=train_test_split(df.drop(columns='Survived'),df['Survived'],test_size=0.2,random_state=42)
```

```python
x_train.head()
```

```python
# imputer Transformer
trf1=ColumnTransformer(
    [
        ('a',SimpleImputer(strategy='mean'),[2]),
        ('b',SimpleImputer(strategy='most_frequent'),[6])
    ]
    ,remainder='passthrough'
)
```

```python
# OneHotEncoding
trf2=ColumnTransformer([
    ('sex_e',OneHotEncoder(sparse_output=False,handle_unknown='ignore'),[1,6])
]
                      ,remainder='passthrough')
```

```python
# Applying MinMaxScaler to all
trf3=ColumnTransformer([
    ('scale',MinMaxScaler(),slice(0,10))
])
```

```python
# Feature Selection
trf4=SelectKBest(score_func=chi2,k=9)
```

```python
# Train the Model
clf=DecisionTreeClassifier()
```

```python
pipe=make_pipeline(
    trf1,
    trf2,
    trf3,
    trf4,
    clf
)
```

```python
pipe.fit(x_train,y_train)
```

```python
result=pipe.predict(x_test)
```

```python
result
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,result)
```

```python
# Cross Validation using cross_val_score
from sklearn.model_selection import cross_val_score
cross_val_score(pipe,x_train,y_train,cv=9,scoring='accuracy').mean()
```
