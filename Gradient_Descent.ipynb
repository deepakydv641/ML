```python
from sklearn.datasets import make_regression
import numpy as np
```

```python
x,y=make_regression(n_samples=4,n_features=1,n_informative=1,n_targets=1,noise=80,random_state=13)
```

```python
x
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x,y)
```

```python
slope=lr.coef_
```

```python
plt.scatter(x,y)
plt.plot(x,lr.predict(x),color="red",label='original')
```

```python
b=200

```

```python
y_pred=slope*x+b
```

```python
plt.scatter(x,y)
plt.plot(x,lr.predict(x),color="red",label='original')
plt.plot(x,y_pred,color="green",label='original')
```

```python
step_size=0.01*(-2*np.sum(y-slope*x-b))
b=b-step_size
```

```python
y_pred1=slope*x+b
```

```python
plt.scatter(x,y)
plt.plot(x,lr.predict(x),color="red",label='original')
plt.plot(x,y_pred,color="green",label='original')
plt.plot(x,y_pred1,color="green",label='original')
```

This is How gradient descent works,as u can see the line getting closer to best fit line in every iteration ,means GD is trying or reducing the error caused by model each and every time

Doing all the stuff that i have done above ,now i will do all those steps by the help of loop

```python
b=-10
fact=0.01
m=slope
epochs=10
plt.plot(x,lr.predict(x),color="red")
for i in range(epochs):
    step_size=fact*(-2*np.sum(y-slope*x-b))
    b=b-step_size
    y_pred=slope*x+b
    plt.plot(x,y_pred)
plt.scatter(x,y)
```
