Now i will make my own class by using Gradient descent and that will be same as the SGDREGRESSOR class implemented by Gradient Descent in SKLEARN,That will be amazing

```python
class GDRegressor:
    def __init__(self,learning_rate,epochs):
        self.m=100
        self.b=0
        self.lr=learning_rate
        self.epochs=epochs
    def fit(self,x_train,y_train):
        for i in range(self.epochs):
            step_size_b=-2*(np.sum((y_train-self.m*x_train-self.b)))/len(x_train)
            step_size_m=-2*(np.sum((y_train-self.m*x_train-self.b)*x_train))/len(x_train)
            self.b=self.b-self.lr*step_size_b
            self.m=self.m-self.lr*step_size_m
    def predict(self,x_test):
        return self.m*x_test+self.b
```

```python
gd=GDRegressor(0.001,9)
gd.fit(x_train,y_train)
```

```python
res=gd.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,res)
```
