```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import KBinsDiscretizer
from sklearn.compose import ColumnTransformer
```

```python
path=r"C:\Users\deepa\Downloads\train.csv"
d=pd.read_csv(path)
```

```python
d.head()
```

```python
df=d[['Age','Fare','Survived']]
```

```python
df.head()
```

```python
df.shape
```

```python
df=df.dropna()
```

```python
x=df.iloc[:,0:2]
y=df.iloc[:,-1]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_test
```

```python
clf=DecisionTreeClassifier()
clf.fit(x_train,y_train)
```

```python
res=clf.predict(x_test)
```

```python
res
```

```python
accuracy_score(y_test,res)
```

```python
# cal accuracy using cross val score 
np.mean(cross_val_score(clf,x,y,cv=100,scoring='accuracy'))
```

```python
# here we have created the object of KbinsDiscretizer
kbin_age=KBinsDiscretizer(n_bins=15,encode='ordinal',strategy='quantile')
kbin_fare=KBinsDiscretizer(n_bins=15,encode='ordinal',strategy='quantile')
```

```python
trf=ColumnTransformer([
    ('a',kbin_age,[0]),
    ('b',kbin_fare,[1])
])
```

```python
x_train_trf=trf.fit_transform(x_train)
```

```python
x_test_trf=trf.transform(x_test)
```

```python
trf.named_transformers_['a'].bin_edges_
```

```python
clf=DecisionTreeClassifier()
clf.fit(x_train_trf,y_train)
```

```python
res2=clf.predict(x_test_trf)
```

```python
accuracy_score(y_test,res2)
```

```python
x_trf=trf.fit_transform(x)
```

```python
np.mean(cross_val_score(clf,x_trf,y,cv=5,scoring='accuracy'))
```
