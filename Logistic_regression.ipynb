```python
from sklearn.datasets import make_classification
import numpy as np
import matplotlib.pyplot as plt
```

```python
x,y=make_classification(n_samples=100,n_features=2,n_informative=1,n_redundant=0,n_classes=2,class_sep=20,n_clusters_per_class=1,random_state=89,hypercube=False)
```

```python
y.shape
```

```python
plt.scatter(x[:,0],x[:,1],c=y)
plt.show()
```

```python
def step(x):
    if x>0:
        return 1;
    else:
        return 0;
```

```python
def perceptron(x,y):
    epochs=10000;
    lr=0.1;
    weights=np.ones(3)
    x=np.insert(x,0,1,axis=1)
    for i in range(epochs):
        j=np.random.randint(0,100)
        y_hat=step(np.dot(weights,x[j]))
        weights=weights+lr*(y[j]-y_hat)*x[j]
    return weights[0],weights[1:]
        
```

```python
intercept_,coef_=perceptron(x,y)
```

```python
intercept_
```

```python
coef_
```

```python
m=-coef_[0]/ coef_[1]
c=-intercept_/ coef_[1]
```

```python
c
```

```python
x_input = np.linspace(-3,3,100)
y_input = m*x_input + c
```

```python
plt.figure(figsize=(10,6))
plt.plot(x_input,y_input,color='red',linewidth=3)
plt.scatter(x[:,0],x[:,1],c=y,cmap='winter',s=100)
plt.ylim(-3,2)
```

This line is predicted(calculated) by perceptron trick of logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
lr.fit(x,y)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
m1=-lr.coef_[0][0]/lr.coef_[0][1]
c1=-lr.intercept_/lr.coef_[0][1]
```

```python

line = m1*x_input + c1
```

```python
plt.figure(figsize=(10,6))
plt.plot(x_input,y_input,color='red',linewidth=3)
plt.plot(x_input,line,color='black',linewidth=3)
plt.scatter(x[:,0],x[:,1],c=y,cmap='winter',s=100)
plt.ylim(-3,2)
```

this black line is more accuaret than the red line(perceptron Trick line) so this is the main disadvantage of the perceptron trick

the black line is more better becoz it is more symmetrical for the groups|
