```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
path=r"C:\Users\deepa\Downloads\placement (1).csv"
```

```python
df=pd.read_csv(path)
```

```python
df.sample(5)
```

```python
df.isna().sum()
```

```python
plt.scatter(df['cgpa'],df['package'])
plt.xlabel("CGPA")
plt.ylabel("Package (lpa)")
df.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
x=df.iloc[:,0].values.reshape(200,1)
y=df.iloc[:,-1].values.reshape(200,1)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train.shape
```

```python
lr.fit(x_train,y_train)
res=lr.predict(x_test)
```

```python
res
```

from sklearn.metrics import r2score

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,res)
```

```python
plt.scatter(df['cgpa'],df['package'])
plt.plot(x_test,res,color="red")
plt.xlabel("CGPA")
plt.ylabel("Package (lpa)")
df.shape
```
