```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
from sklearn.impute import SimpleImputer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.tree import DecisionTreeClassifier
```

```python
# Take a Raw string by using r before the path of the file
path=r"C:\Users\deepa\Downloads\train.csv"
```

```python
d=pd.read_csv(path)
```

```python
d.head(2)
```

```python
#drops the colums which are of no use
df=d.drop(columns=['PassengerId','Name','Ticket','Cabin'])
```

```python
df.head()
```

```python
#checks which col have how many missing values
df.isna().sum()
```

```python
#Does train test split
x_train,x_test,y_train,y_test=train_test_split(df.drop(columns='Survived'),df[['Survived']],test_size=0.2,random_state=42)
```

```python
x_train
```

```python
x_train['Age']
```

```python
x_train['Embarked']
```

```python
# performed imputation on col having missing vales
si_age=SimpleImputer()
si_ed=SimpleImputer(strategy='most_frequent')
x_train_age=si_age.fit_transform(x_train[['Age']])
x_train_ed=si_ed.fit_transform(x_train[['Embarked']])
x_test_age=si_age.transform(x_test[['Age']])
x_test_ed=si_ed.transform(x_test[['Embarked']])
```

```python
x_train_ed
```

```python
# OneHotEncoding on Sex and Embarked
ohe_sex=OneHotEncoder(sparse_output=False,handle_unknown='ignore')
ohe_ed=OneHotEncoder(sparse_output=False,handle_unknown='ignore')
x_train_sex=ohe_sex.fit_transform(x_train[['Sex']])
x_train_ohe_ed=ohe_ed.fit_transform(x_train_ed)
x_test_sex=ohe_sex.transform(x_test[['Sex']])
x_test_ohe_ed=ohe_ed.transform(x_test_ed)
```

```python
x_train_sex
```

```python
x_train_ohe_ed
```

```python
# Now I need to CONCATENATE
x_train_rem=x_train.drop(columns=['Sex','Age','Embarked'])
x_test_rem=x_test.drop(columns=['Sex','Age','Embarked'])

# here i will Concatenate

x_train_new=np.concatenate((x_train_rem,x_train_age,x_train_sex,x_train_ohe_ed),axis=1)
x_test_new=np.concatenate((x_test_rem,x_test_age,x_test_sex,x_test_ohe_ed),axis=1)
```

```python
x_train_new
```

```python
x_test_rem
```

```python
# here we have trained our model
clf=DecisionTreeClassifier()
clf.fit(x_train_new,y_train)
```

```python
y_pred=clf.predict(x_test_new)
```

```python
y_pred
```

```python
# Here we calculate the accuracy of our model
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
