```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.impute import KNNImputer,SimpleImputer
from sklearn.linear_model import LogisticRegression
```

```python
path=r"C:\Users\deepa\Downloads\train.csv"
d=pd.read_csv(path,usecols=['Age','Pclass','Fare','Survived'])
```

```python
d.head()
```

```python
d.isna().sum()
```

```python
x=d.drop(columns=['Survived'])
y=d['Survived']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_test.shape
```

```python
si=SimpleImputer()
x_train_si=si.fit_transform(x_train)
x_test_si=si.transform(x_test)

x_comp=si.transform(x)
```

```python
lr=LogisticRegression()
lr.fit(x_train_si,y_train)
```

```python
res=lr.predict(x_test_si)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,res)
```

```python
from sklearn.model_selection import cross_val_score
np.mean(cross_val_score(lr,x_comp,y,cv=5,scoring='accuracy'))
```

```python
# in knn there is n_neighbour and weights attributes are there,weights can take uniform or distance ,uniform is just normal and in distance we multiply the value by their inverse of distance means nearer value will share more proportion in the answer

knn=KNNImputer()

x_train_knn=knn.fit_transform(x_train)
x_test_knn=knn.transform(x_test)

x_comp_knn=knn.transform(x)
```

```python
lr=LogisticRegression()
lr.fit(x_train_knn,y_train)
```

```python
result=lr.predict(x_test_knn)
```

```python
accuracy_score(y_test,result)
```

```python
np.mean(cross_val_score(lr,x_comp_knn,y,cv=5,scoring='accuracy'))
```
