```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt

path=r"C:\Users\deepa\Downloads\titanic_toy.csv"

d=pd.read_csv(path)
```

```python
d.isna().mean()*100
```

```python
x=d.drop(columns=['Survived'])
y=d['Survived']
```

Using sklearn ,u can also do by using pandas fillna

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train.shape,x_test.shape
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
trf=ColumnTransformer([
    ('a',SimpleImputer(strategy='median'),['Age']),
    ('b',SimpleImputer(strategy='mean'),['Fare'])],remainder='passthrough')
```

```python
x_train_trf=trf.fit_transform(x_train)
trf.transform(x_test)
```
