```python
from sklearn.feature_selection import SelectKBest,chi2
from sklearn.pipeline import make_pipeline
```

```python
df=d.drop(columns='id')
```

```python
df.head()
```

```python
x_train,x_test,y_train,y_test=train_test_split(df.drop(columns='stroke'),df['stroke'],test_size=0.2,random_state=42)
```

```python
x_train.head()
```

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import MinMaxScaler
```

```python
# Simple Imputer Transformer
trf1=ColumnTransformer([
    ('a',SimpleImputer(strategy='mean'),[8])],remainder='passthrough')
```

```python
# one Hot Encoding
trf2=ColumnTransformer([
    ('b',OneHotEncoder(drop='first',sparse_output=False,handle_unknown='ignore'),[0,4,5,6,9])],remainder='passthrough')
```

```python
trf3=ColumnTransformer([
    ('c',MinMaxScaler(),slice(0,16))])
```

```python
# feature Selection
trf4=SelectKBest(score_func=chi2,k=5)
```

```python
trf5=LogisticRegression()
```

```python
pipe=make_pipeline(trf1,trf2,trf3,trf4,trf5)
```

```python
pipe.fit(x_train,y_train)
```

```python
result=pipe.predict(x_test)
```

```python
result
```

```python
accuracy_score(result,y_test)
```
